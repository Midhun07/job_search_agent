```python
import os
from google import genai

# Remove the decorator and pytest.skip
def get_client():
    api_key = os.getenv("GEMINI_API_KEY")
    if not api_key:
        raise ValueError("GEMINI_API_KEY not set in environment or .env")
    return genai.Client(api_key=api_key)

# Now you can call it manually without errors!
client = get_client() 
print(list(client.models.list()))
```

```python
from google.genai.errors import ClientError

try:
    response = client.models.generate_content(
        model="gemma-4-31b-it",
        contents="What is the capital of Sri Lanka?",
    )
except ClientError as e:
    print(f"ClientError: {e}")
    raise
text = response.text.strip()
# assert text, "Response text should not be empty"
# assert "PONG" in text.upper()
```
